###  Q6 Median Value of Horsepower

1. Find the median value of the horsepower column in the dataset.
2. Next, calculate the most frequent value of the same horsepower column.
3. Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step.
4. Now, calculate the median value of horsepower once again.
 - Has it changed?

```python
#Q6.1 - mean of horsepower column
median_before = df['horsepower'].median().item()
print('Q6 Part 1:', median_before)
```

```python
#Q6.2 max frequency of horsepower column
# max freq = mode
mode_horsepow = df['horsepower'].mode().item()
mode_horsepow
print('Q6.2:', mode_horsepow)
```

```python
#Q6.3 - Fill Null elements
n_null = df['horsepower'].isnull().sum().item() 
print('Total of Null elements:', n_null)
```

```python
#fill Null elements with max freq values 
df['horsepower'] = df['horsepower'].fillna(mode_horsepow) # 'mode_horse' has to be integer - 'fillna()' does not accept float
df
```

```python
#check null values
df['horsepower'].isnull().sum().item()
```

```python
#Q6.4 - new median value
median_after = df['horsepower'].median().item()
print('Q6 Part 4:', median_after)
```

```python
#Q6.5 - Has it changed
# Uses boolean
print('Has it changed?', 
      'No' if median_before == median_after 
      else 'Yes, increased' if median_after - median_before > 0 
      else 'Yes, decreased')
```

### Q7 -  Sum of Weights

1.  Select all the cars from Asia
2. Select only columns vehicle_weight and model_year
3. Select the first 7 values
4. Get the underlying NumPy array. Let's call it X.
5. Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
6. Invert XTX.
7. Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
8. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
9. What's the sum of all the elements of the result?
Note: You just implemented linear regression. 

```python
df
```

```python
#Q7.1 - Select Asia Cars
asia = df[df['origin'] == 'Asia']
asia
```

```python
#Q7.2 Select 2 columns - 'vehicle_weight' and 'model_year'
asia[['vehicle_weight', 'model_year']]
```

```python
#Q7.3 - get first 7 values (rows)
asia[['vehicle_weight', 'model_year']].iloc[:7].values.round(2)
```

```python
#Q7.4 - Get the array shape of X
X = asia[['vehicle_weight', 'model_year']].iloc[:7].values.round(2)
X.shape
print('X array shape is:', X.shape)
```

```python
#Q7.5 - matrix mult between tranpose X and X
XTX = X.T @ X
XTX
print('Matrix-Multiplication of X^T and X is:', XTX)
# note: title of print() mix with array distorted
```

```python
#Q7.6 - Invert a matrix
XTX_inv = np.linalg.inv(XTX)
print('XTX Inverse is:', XTX_inv)
```

```python
#Q7.7 - create array y with specified values
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
#Q7.8 - Multiply inverse matrix with its origin, and multiply y, call w
w = XTX_inv @ X.T @ y
w
```

```python
#Q7.9 - sum of matrix w - done
sum_w = w.sum().round(3).item()
print('The sum of all elements in w matrix is:', sum_w)
```
