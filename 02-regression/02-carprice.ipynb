## NOTE 
Goal: Look at values    - for val in xi:

Need matching items from 2 lists - for j in range(len(xi)):  (99 % of ML)

```python
w0 = 0
```

```python
def linear_regression(xi):
    n = len(xi) # len(xi) for size of feature matrix, assign to n
    pred = w0 #prediction w0 into variable 'pred'

    # create a loop for j in range of n = len(xi) = 3 = n
    # range(n) just gives you the indices 0 to n-1 - in this case 3 (index 2) - 0 to 2
    # so you can safely access every weight and every feature, no matter how many there are.
    for j in range(n):
        pred = pred + w[j] * xi[j]
    return pred
    
```

```python
xi = [453, 11, 86]
w = [1, 1, 1]
w0 = 7.5

# OLD WAY 'in range()' method
def old_way(xi):
    pred = w0
    for j in range(len(xi)):
        pred += w[j] * xi[j]
    return pred
old_way(xi)
```

```python
# enumerate() WAY
# cheat code ends 90 % index headaches forever.
# gives both the index & value - ML 99% of time

def enu_way(xi):
    pred = w0
    for j, val in enumerate(xi):
        pred += w[j] * val # j = index, value = actual number
    return pred
enu_way(xi)
```

```python
# xi (log feature matrix) to Xi
# dot product
def dot(xi, w):
    res = 0.0
    for j, val in enumerate(xi):
        res += w[j] * val
    return res
```

## 2.7 Training A Linear Regression Model

```python
# coming up with weights 'w' 
# g(X) = Xw =~ y
# use X^-1:    X^(-1)*X = 1 (Gram matrix)  
# (X^-1 * X) * w = y ---->  (1)w = X^(-1)Y
# where 1 is I - identity matrix
# (I)w = X^(-1)Y
```

```python
X = [
    [148, 24, 1385],
    [132, 25, 2031],
    [453, 11, 86],
    [158, 24, 185],
    [172, 25, 201],
    [413, 11, 86],
    [38, 54, 185],
    [142, 25, 431],
    [453, 31, 86]
]

X = np.array(X)
X
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
train_linear_regression(X, y)
```

## 2.9 Root Mean Squared Error

```python
# mrse = sqrt( 1/m *  (summation_i=1_top=m, (g(xi) - yi)^2))
# where yi = actual price, g(xi) = prediction
```

```python
y, y_train # y = actual, actual value, y_train is np.log1p() of value
```

```python
def rmse(y, y_pred): 
       # recall 'def' a function - 'y', and 'y_pred' are just placeholders
       # gets replaced when function is called 
       se = (y - y_pred) ** 2 # error value, e, to square error, se
       mse = se.mean() #  mean of se, mse
       return np.sqrt(mse) # sqrt(mse) using 'np.sqrt()'
```

## 2.13 Regularization

```python
def train_linear_regression_reg(X, y, r=0.0):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    reg = r * np.eye(XTX.shape[0])
    XTX = XTX + reg

    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]
```

```python
X_train = prepare_X(df_train)
```

```python
for r in [0, 0.001, 0.01, 0.1, 1, 10]:
    w_0, w = train_linear_regression_reg(X_train, y_train, r=r)
    print('%5s, %.2f, %.2f, %.2f' % (r, w_0, w[13], w[21]))
```
